```python
import pandas  as pd
training = pd.read_csv("./titanic-train.csv")
training.info()
```

```python
training.head()
```

```python
training["Gender"] = training["Gender"].apply(lambda toLabel: 0 if toLabel == 'male' else 1)
```

```python
training.head()
```

```python
training["Age"].fillna(training["Age"].mean(), inplace = True)
training.info()
```

```python
training
```

```python
training.describe()
```

```python
#Entrenar y calificar el modelo de arbol de decisiones
#Se crea un objeto de arreglo con la variable que será el objetivo del modelo
y_target = training["Survived"].values
```

```python
#Cree el arreglo de los valores que será la entrada para el modelo
columns = ["Fare","Pclass","Gender","Age","SibSp"]
X_input = training[list(columns)].values
```

```python
#Se crea el modelo programado
from sklearn import tree

clf_train = tree.DecisionTreeClassifier(criterion="entropy", max_depth=3)
clf_train = clf_train.fit(X_input, y_target)
```

```python
clf_train.score(X_input,y_target)
```
